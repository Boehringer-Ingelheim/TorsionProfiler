# Calculate TorsionProfiles:
This Notebook will give a quick starter on how to use torsion_profiler form pyton!



## Get test Data

There are many dataset with small molecules in the pyQMLprop repository contained!

```python
from torsion_profiler.data import torsion_profiles
print("\n".join(torsion_profiles.data_sets))
```

All of them have different amount of molecules and complexities for the different molecules.
We recommend with starting the ToyTestSet or the RowleyBiarylSet as they are rather small and not too complex molecules.

You can import the datasets as follows:

```python
import ast
from torsion_profiler.data.torsion_profiles import ToyTestSet

#input Test data 
mols = ToyTestSet.get_mols()

print("got Molecules: ", len(mols))

mol = mols[0]
torsion_atom_ids = ast.literal_eval(mol.GetProp("torsion_atom_ids"))
mol
```

```python
from torsion_profiler.visualize.visualize_tp_3d import show_3d_mol
show_3d_mol(mol)
```

## Calculate Torsion Profiles
Next we are going to calculate a Torsion Profile directly in the jnb.

### Calculate single Torsion

You can switch the calculators by simpling putting a different class there!


```python
# Setup machinery
from torsion_profiler import TorsionProfiler
from torsion_profiler.engines import Mmff94Calculator, OffCalculator, AniCalculator

calculator = Mmff94Calculator()
tp = TorsionProfiler(calculator=calculator, n_measurements=20)

# Calculate
df = tp.calculate_torsion_profile(mol=mol, 
                                 torsion_atom_ids=torsion_atom_ids)

df
```

# Plot the Torsion Profile

```python
from torsion_profiler.visualize import plot_torsion_profile
fig = plot_torsion_profile(tp_df=df, title_prefix="Butan")
```

# Let's have a fancy 3D widget!

```python
from torsion_profiler.visualize import show_3d_mol
from torsion_profiler.utils.metrics import get_simple_energy_classifications

torsion_atom_ids = df["torsion_atom_ids"].iloc[0]
colors = get_simple_energy_classifications(df["rel_potential_energy"])
show_3d_mol(mol=df["ROMol"], highlight_atoms=torsion_atom_ids, color=colors)
```
